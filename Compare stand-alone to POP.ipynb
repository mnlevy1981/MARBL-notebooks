### Did we correctly capture POP forcing data?

Hopefully the dataset created from [Forcing Data (with xarray).ipynb](https://github.com/mnlevy1981/MARBL-notebooks/blob/master/Forcing%20Data%20(with%20xarray).ipynb) produces the same MARBL output in its five columns as we find in the POP datasets.

```python
import xarray as xr
import numpy as np
import physics # local file with helper functions
import shared_forcing

import matplotlib
import matplotlib.pyplot as plt

grid_pts, ncol = shared_forcing.set_grid_points()
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (18.0, 10.0); # Large figures
```

```python
ds_pop = shared_forcing.read_pop_data_set()
ds_marbl = shared_forcing.read_marbl_data_set()
```

```python
comp_list = []
for var in ds_marbl:
    if var not in ds_pop:
        print("{} is in MARBL data set but not POP".format(var))
    else:
        comp_list.append(var)
```

```python
for var in comp_list:
    if True:#var.startswith('TEND_'):
        print('{}\n---'.format(var))
        for n, pt in enumerate(grid_pts.keys()):
            pop_data = ds_pop[var].isel(time=0, nlat=grid_pts[pt]['lat'], nlon=grid_pts[pt]['lon']).data
            marbl_data = ds_marbl[var].isel(num_cols=n).data
            abs_err = np.nanmax(np.abs(pop_data - marbl_data))
            rel_err = abs_err/np.nanmax(pop_data)
            print('Col {}\n  * max diff: {}\n  * rel diff: {}'.format(n+1, abs_err, rel_err))
            #print('Col {}\n  * pop data: {}\n  * marbl data: {}\n  * max diff: {}\n  * rel diff: {}'.format(n+1, pop_data, marbl_data, abs_err, rel_err))
        print('')
```
